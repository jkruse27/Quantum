# Fundamentos Para o Uso do Qiskit

Neste notebook nós revisaremos todo o básico para o uso do ambiente Qiskit da IBM, incluindo como criar, visualizar e simular circuitos. 

### Parte 1 - O que será necessário

Para podermos utilizar as funções de criação, visualização e simulação de circuitos quânticos será necessário importar a seguinde biblioteca (tutorial para sua instalação no github oficial https://github.com/Qiskit/qiskit).

```python
#importa tudo da biblioteca qiskit
from qiskit import *
 
```

### Part 2 - Criando e Visualizando Circuitos Quânticos

Agora que importamos todo o necessário estamos prontos para começar. Para criar um circuito quântico, existe o método QuantumCircuit() que recebe como parâmetro o numero de registradores. Esses registradores podem ser pensados como as linhas do circuito, significando que cada um deles guarda um qubit que poderá ser modificado ou ligado à outros qubits atravez de portas quânticas ("quantum gates"). Então, vamos criar um circuito. É importante saber que, por padrão, os qubits começam com o valor $ \left| 0 \right\rangle $, de forma que, se medirmo-os diretamente obteriamos em 100% dos casos$ \left| 0 \right\rangle $.

```python
#Creating a quantum circuit holding 2 registers
circuit = QuantumCircuit(2)
```

For visualizing the circuit, we use QuantumCircuits function draw(). As seen below, the circuit has 2 lines (as it has 2 registers) and there is still no logic gates applied to them.

```python
circuit.draw()
```

### Part 3 - Simulation

For this part, we'll be simulating the results obtained from the circuit as if it was run from a real quantum computer. It is worth noting that IBM's plataform allows for users to actually simulate their programs in real quantum computers, but for now we will be working with normal computers simulation.

We are going to use Aer's statevector simulation. This simulation returns a vector of the states of the system containing $2^{n}$ dimensions, n being the number of qubits in the system. Because of this, this method is not recommended for large simulations as it is exponencial and so will get really big really fast. For this, you'll need to create a backend, in which you will run your job and from which you will get your results. Doing all of this is really simple, as shown in the cell below. The state vector will have in each position a complex number indicating the probability of each state (in this 2 register case: 00, 01, 10, 11). It is worth noting that the state vector simulation returns the ideal results for the circuit, without considering noise interference or decoherence or anyother thing.

```python
# Creates the backend with state vector simulator
backend = Aer.get_backend('statevector_simulator')

# Runs the circuit on this backend
# When you execute your circuit on the backend, it return a job
# This job contains the status and the results of the execution (job.status(), jos.result()) 
job = execute(circuit, backend)

# What we want now is the results
results = job.result()

# With the results we can get the state vector using get_statevector(), 
# which receives the circuit as a parameter (it has other optional parameters too)
state_vector = results.get_statevector(circuit)

print(state_vector)
```

For a more pleasing view, it is possible to plot this state vector.

```python
# Plots the state vector for us
plot_state_city(state_vector)
```

As we've seen before, because the initial states for the qubits is $ \left| 0 \right\rangle $ and we haven't done anything to them, their final state will be $ \left| 0 \right\rangle $ to all the time, as is shown in the plot and in the state vector. We know as well that the sum of the squares of the probabilities (in this case, as we are dealing with complex numbers, the probability is given by $\sqrt{Re^{2} + Im^{2}}$) is equal to 1. It's also important to add that the number most to the left in the plot represents the register 1 and the one most to the right represents the register 0 (like a binary number, the least significant register is most to the right).

### Part 4 - Quantum Gates

   Now we know how to create and display our quantum circuits, but they still don't do anything. For them to start being usefull, we will need to add quantum gates. These gates work by changing the possibility of a qubits state or by linking (entangling) to qubits states together so that one qubits state influences the other. It is also important to add that all quantum gates are reversible, which mean that, if applied twice consecutevely, they return to the original state. In Qiskit, the registers order in the circuit will be the order in which they are added, so that the first one added will be most to the left and the last one most to the right.

#### Hadamard Gate 

The first gate we will take a look at is the hadamard gate, denominated as H. This gate works in a single qubit and it puts the qubit in a state of superposition (in this case it has an equal probability of being in a state one or in a state zero). Later on we will see a more thorough explanation, but for now that's enough for understanding the basics. Now let's take a look at how it is implemented. In order to insert a Hadamard gate to a circuit register, it is necessary to use the methos QuatumCircuit.h(), which receives the number of the register it will be applied to as a parameter. Below you'll see the code implemented showing the gate and its effect on the registers output.

```python
# Adding a Hadamard gate to register 0
circuit.h(0)

# Drawing the circuit
circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

As expected, when we add a Hadamard gate to the first qubit, it's probability of being in state $ \left| 0 \right\rangle $ and $ \left| 1 \right\rangle $ is equal, as shown in the graph, where the qubit to the left is always 0 and the one to the right is equaly likely to be 1 or 0. Now, if we apply the Hadamard gate again to the same qubit without doing anything to it between the applications, let's see what happens.

```python
circuit.h(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

The circuit returns to its original state, which is to be expected as we saw quantum gates are reversible.

#### Not Gate

Now we are going to take a look at the not gate, denoted by x. The not gate is applied to a single qubit and what it does is it flips the qubits state, so if it was a $ \left| 1 \right\rangle $ it becomes a $ \left| 0 \right\rangle $ and vice-versa. To apply this gate to the circuit we use the QuantumCircuit.x() method, that receives the number of the register it will be applied to.

```python
circuit = QuantumCircuit(2)

# Applying the not gate to the 0 register
circuit.x(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

As expected, register's 0 state was flipped from 0 to 1. Now let's see what happens if we apply the not gate after the Hadamard gate.

```python
circuit = QuantumCircuit(2)

circuit.h(0)
circuit.x(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

Nothing happened, it is as if the not gate wasn't even there. In further notebooks we will see in more detail why that is so, but for now let's just say that, as register 0's states was equaly likely, inverting them ended up doing nothing. Now what would happen if we applied the not gate first.

```python
circuit = QuantumCircuit(2)

circuit.x(0)
circuit.h(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

This time something really strange happened. We still don't have the knowledge to understand this but soon enough we will. Just to end this let's show that the not gate is inversible too.

```python
circuit = QuantumCircuit(2)

circuit.x(0)
circuit.x(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

As foreseen, the result is the same as if we hadn't used the gate.

#### Conditional Not Gate

The last gate we are seeing in this note book is the conditional not gate, denoted by cx. This gate works on two qubits, one control bit and one target bit. If the control bit is $ \left| 1 \right\rangle $, the target bit state is flipped else nothing happens. To add this gate to the circuit you use the QuantumCircuit.cx() method, which receives 2 parameters, the first is the control register number and the second the target register number. The result after the use of this gate is a entangled state between both the qubits.

```python
circuit = QuantumCircuit(2)

circuit.x(0)

# Entangles the qubits with 0 as the control and 1 as the target
circuit.cx(0, 1)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

As we can see, because we flipped the register 0, it became the state $ \left| 1 \right\rangle $. Because it was in the state $ \left| 1 \right\rangle $, it flipped the register 1 from $ \left| 0 \right\rangle $ to $ \left| 1 \right\rangle $, resulting in the state 11 100% of the time. If we applied the not gate after the link what would have happened?

```python
circuit = QuantumCircuit(2)

circuit.cx(0, 1)
circuit.x(0)

circuit.draw()
```

```python
# Plotting to see it's effects
backend = Aer.get_backend('statevector_simulator')
job = execute(circuit, backend)
results = job.result()
state_vector = results.get_statevector(circuit)
print(state_vector)
plot_state_city(state_vector)
```

Because the register 0 was in state $ \left| 0 \right\rangle $ when we applied cx, it didn't switch register 1's state. Thus, the final state is 01 because register 1 was not changed and register 0 was flipped when the not gate was used afterwards.
