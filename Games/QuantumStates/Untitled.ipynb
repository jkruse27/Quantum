# The Qubit

The qubit can be represented as a vector:
\begin{bmatrix}
\alpha \\
\beta
\end{bmatrix}
or $$\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $$
Applying a gate G in the qubit is the equivalent of multiplying the state $\left| \psi \right\rangle $ by the matrix representing the gate (G$\left| \psi \right\rangle$).

# Gates
## Gate X

This gate flips the state of the qubit 180 degrees on the X axis. It can be represented by the matrix:
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
, taking the $\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $ state to the $\beta \left| 0 \right\rangle+\alpha\left| 1\right\rangle $ state.

## Gate Y

This gate flips the state of the qubit 180 degrees on the Y axis. It can be represented by the matrix:
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix}
, taking the $\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $ state to the $-i\beta \left| 0 \right\rangle+i\alpha\left| 1\right\rangle $ state.

## Gate Z

This gate flips the state of the qubit 180 degrees on the Z axis. It can be represented by the matrix:
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
, taking the $\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $ state to the $\alpha \left| 0 \right\rangle-\beta\left| 1\right\rangle $ state.

## Gate H

This gate puts the qubit in a superposition state. It can be represented by the matrix:
$$\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
$$
, taking the $\left| 0 \right\rangle$ state to the $\left| + \right\rangle$ state and the $\left| 1 \right\rangle$ state to the $\left| - \right\rangle$.($\left| + \right\rangle$ = $\frac{\left| 0 \right\rangle + \left| 1 \right\rangle}{\sqrt{2}}$ and $\left| - \right\rangle$ = $\frac{\left| 0 \right\rangle - \left| 1 \right\rangle}{\sqrt{2}}$)

## Gate T

This gate has a $\phi=\frac{\pi}{4}$. It can be represented by the matrix:
\begin{bmatrix}
1 & 0 \\
0 & e^{\frac{i\pi}{4}}
\end{bmatrix}
, taking the $\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $ state to the $\beta \left| 0 \right\rangle+\alpha\left| 1\right\rangle $ state.

## Gate S

This gate does a quarter-turn along the bloch sphere. It can be represented by the matrix:
\begin{bmatrix}
1 & 0 \\
0 & i
\end{bmatrix}
, taking the $\alpha \left| 0 \right\rangle+\beta\left| 1\right\rangle $ state to the $\beta \left| 0 \right\rangle+\alpha\left| 1\right\rangle $ state.
